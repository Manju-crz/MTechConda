```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df['Month'] = pd.to_datetime(df['Month'], format='%d-%m-%Y')
df = df.rename(columns={'Month': 'ds', 'Avg_sunspot_count': 'y'})
```

```python
train_df = df[df['ds'] < '2001-01-01']
test_df = df[(df['ds'] >= '2001-01-01') & (df['ds'] <= '2010-12-01')]
```

```python
model = Prophet(yearly_seasonality=True)
model.fit(train_df)
```

```python
future = model.make_future_dataframe(periods=120, freq='MS')
forecast = model.predict(future)
```

```python
forecast_test = forecast[(forecast['ds'] >= '2001-01-01') & (forecast['ds'] <= '2010-12-01')]
merged = forecast_test[['ds', 'yhat']].merge(test_df, on='ds')
```

```python
mae = mean_absolute_error(merged['y'], merged['yhat'])
rmse = np.sqrt(mean_squared_error(merged['y'], merged['yhat']))
mape = np.mean(np.abs((merged['y'] - merged['yhat']) / merged['y'])) * 100
```

```python
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(merged['ds'], merged['y'], label='Actual')
plt.plot(merged['ds'], merged['yhat'], label='Predicted')
plt.title('Sunspot Count Forecast (20012010)')
plt.xlabel('Date')
plt.ylabel('Avg Sunspot Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
test_df = pd.read_csv("test.csv")  # Replace with your file name
test_df['ds'] = pd.to_datetime(test_df['Month'], format='%d-%m-%Y')

# STEP 5: Predict using the trained model
forecast = model.predict(test_df[['ds']])

# STEP 6: Prepare output with required column
output_df = forecast[['ds', 'yhat']]
output_df = output_df.rename(columns={'ds': 'Month', 'yhat': 'Avg_sunspot_count'})

# STEP 7: Save to a new CSV
output_df.to_csv("sunspot_forecast_test_output.csv", index=False)

print(" Prediction saved to 'sunspot_forecast_test_output.csv'")
```
