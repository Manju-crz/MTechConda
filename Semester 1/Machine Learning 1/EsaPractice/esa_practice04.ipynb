```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Compute VIF for each feature
def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = df.columns
    print("features list", vif_data["Feature"])
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    print("VIF list\n", vif_data["VIF"])
    return vif_data
calculate_vif(df)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# Generate sample data
np.random.seed(42)
X = np.random.rand(100, 5)
y = 3*X[:, 0] + 2*X[:, 2] + 5*X[:, 4] + np.random.randn(100)

# Define model
lasso = Lasso(max_iter=500)

# Define hyperparameter grid
param_grid = {
    "alpha": [1e-4, 1e-3, 0.01, 0.1, 1, 10, 100]  # Different alpha values to test
}

# Perform GridSearchCV
grid_search = GridSearchCV(lasso, param_grid, cv=5, scoring="neg_mean_squared_error", verbose=1)
grid_search.fit(X, y)
```

```python
# Get best parameters and best score
print("Best Alpha:", grid_search.best_params_["alpha"])
print("Best MSE:", -grid_search.best_score_)

# Train final model with best alpha
best_lasso = Lasso(alpha=grid_search.best_params_["alpha"], max_iter=500)
best_lasso.fit(X, y)

# Print final coefficients
print("Final Coefficients:", best_lasso.coef_)
```
