# Map

```python
import pandas as pd
```

```python
df1 = pd.read_csv("lds1.csv")
```

```python
df1
```

```python
df1.Vol.mean()
```

```python
df1.Vol.quantile()
```

```python
def sales_performance(value) :
    if value <= df1.Vol.quantile(0.33):
        return 'Average Sales'
    elif value >= df1.Vol.quantile(0.66):
        return 'Excellent Sales'
    else :
        return 'Good Sales'
```

```python
df1['Sales_Performance'] = df1.Vol.map(sales_performance)
```

```python
df1['Sales_Performance']
```

```python
df1
```

# Box Plot Boundaries

```python
v1 = list(range(300,400))
print(np.mean(v1))
print(np.min(v1))
print(np.max(v1))

print(np.percentile(v1,0.50))
print(np.percentile(v1,0.25))
print(np.percentile(v1,0.75))
```

```python
plt.boxplot(v1)
```

```python
q2 = np.percentile(v1,0.50)
q1 = np.percentile(v1,0.25)
q3 = np.percentile(v1,0.75)
```

```python
print(q2)
print(q1)
print(q3)
```

```python
iqr = q3-q1
print(q1,q2,q3,iqr)
```

```python
lw = q1 - (1.5 * iqr)
uw = q3 + (1.5 * iqr)
print(lw, uw)
```
