```python
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [1, 1, 1, 2, 2],
    'C': [1, 2, 3, 4, 10]
})

# Compute skewness for each column
skewness = df.skew()
print(skewness)
```

```python
import pandas as pd

# Example DataFrame with negative skew
data = {
    'Column1': [1, 1, 1, 1, 5],
    'Column2': [10, 10, 10, 10, 1]
}

df = pd.DataFrame(data)

# Calculate skewness
skewness = df.skew()
print(skewness)
```

```python
# Kurtosis
"""
It is a statistical measure that defines how heavily the tails of the distribution differ from the normal disrtibution.
It identifies, whether the tails of a given distribution contain extreme values
Thicker the tails are more extreme values in the data
"""

```

```python
df21
```

```python
df21.kurt(numeric_only = True)
```

```python
# Example DataFrame with negative kurtosis
data = {
    'Column1': [1, 2, 3, 1, 2, 0, 1],         # Normal distribution should have kurtosis close to 0
    'Column2': [1, 1, 2, 2, 3, 4, 5]    # Slightly uniform distribution
}

df = pd.DataFrame(data)

# Calculate kurtosis
kurtosis = df.kurt()
print(kurtosis)
```
