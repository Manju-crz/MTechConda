```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
l1 = [3,5,7,2,7,9,3,4,6]
l1
```

### Line Plot

```python
plt.plot(l1)
```

```python
l2 = [-3,5,7,-2,7,9,3,4,6]
plt.plot(l2)
plt.title('My First Plot')
plt.show()
```

```python
plt.plot(l2)
plt.title('My First Plot')
plt.xlabel('Std Indices')
plt.ylabel('Plt Range')
plt.grid(linestyle = '-.', linewidth='0.25', color = 'orange')
plt.show()
```

### Multi-Line Plot

```python
x = [1,2,3,4]
y1 = [80000, 78000, 87000, 98000]
y2 = [75000, 14000, 18000, 80000]
y3 = [45000, 45000, 83333, 80000]
y4 = [47000, 45000, 78000, 45700]
```

```python
plt.plot(x, y1, label = 'Samsung', linewidth = 2.5)
plt.plot(x, y2, label = 'HP', linewidth = 2.5)
plt.plot(x, y3, label = 'Lenovo', linewidth = 2.5)
plt.plot(x, y4, label = 'Asus', linewidth = 2.5)
plt.title('Share Market Price')
plt.xlabel('Days')
plt.ylabel('Price per lot')
plt.grid(linestyle = '-.', linewidth='0.25', color = 'orange')
plt.legend()
plt.show()
```

### Bar Plot

```python
x = np.array([1,2,3,4])
y1 = [80000, 78000, 87000, 98000]
y2 = [75000, 14000, 18000, 80000]
y3 = [45000, 45000, 83333, 80000]
y4 = [47000, 45000, 78000, 45700]
```

```python
plt.bar(x, y1)
```

```python
plt.barh(x, y1)
```

```python
width = 0.10
plt.bar(x, y1, width, label = 'Samsung')
plt.bar(x + width, y2, width, label = 'HP')
plt.bar(x + width + width, y3, width, label = 'Lenovo')
plt.legend()
```

### Pie Plot

```python
l11
```

```python
l12
```

```python
labels=['A','B','C','D']
colors = ['red','blue','yellow','pink']
```

```python
plt.pie(l12, labels = labels, colors = colors)
plt.show()
```

```python
plt.pie(l12, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
def func(pct, allvalues):
    absolute = int(round(pct / 100. * sum(allvalues)))
    return f'{absolute}'

plt.pie(l12, labels=labels, colors=colors, autopct=lambda pct: func(pct, l12))
plt.show()
```

### Exploded pie plot

```python
l12 = [5, 25, 30, 40]
labels = ['A','B','C','D']
colors = ['red','blue','yellow','pink']
explode = [0.2, 0, 0, 0.1]
plt.pie(l12, labels=labels, explode = explode, colors=colors, autopct='%1.1f%%', shadow = True)
plt.axis('equal')
plt.show()
```

### Donut pie plot

```python
l12 = [5, 25, 30, 40]
labels = ['A','B','C','D']
colors = ['red','blue','yellow','pink']
```

```python
l12
```

```python
plt.pie(l12, labels=labels, colors=colors, autopct='%1.1f%%', shadow = True)
circle = plt.Circle( (0,0), 0.5, color = 'white' )
plt.gcf()
plt.gca().add_artist(circle)
plt.show()
```

# Introduction to Seaborn

```python
pip install seaborn
```

```python
import seaborn as sns
```

### Strip plot

```python
df31 = pd.read_csv('lds0.csv')
df31
```

```python
sns.stripplot(x='StockID', y = 'Rev', data = df31)
plt.show()
```

```python
sns.stripplot(x='StockID', y = 'Rev', hue = 'StockID', data = df31)
plt.show()
```

### Violin plot

```python
sns.violinplot(x='StockID', y = 'Rev', hue = 'StockID', data = df31)
plt.show()
```

```python
sns.violinplot(x='StockID', y = 'Rev', hue = 'Started', data = df31, split = True)
plt.show()
```

### Swarm plot

```python
sns.swarmplot(x='StockID', y = 'Rev', hue = 'StockID', data = df31)
plt.show()
```

```python
df32 = pd.read_csv('lds41.csv')
df32
```

```python
sns.swarmplot(x='am', y = 'disp', hue = 'am', data = df32)
plt.show()
```

### Pair plot

```python
sns.pairplot(df32)
plt.show()
```

### Distribution Plot

```python
sns.histplot(df32['disp'])
plt.show()
```

### Count Plot

```python
sns.countplot(x='StockID', data = df31)
plt.show()
```

### Heatmap

```python
sns.heatmap(df31[['Rev','Sector','HQ','StockID']].corr(), annot = True)
plt.show()
```

```python
df31
```

```python
sns.heatmap(df31[['Rev','Sector','HQ','StockID']].corr(), annot = True, linewidth = 0.3)
plt.show()
```
