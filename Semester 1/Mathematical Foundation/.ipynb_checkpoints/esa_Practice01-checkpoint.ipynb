```python
import math

def sin_theta(degrees):
    radians = math.radians(degrees)  # Convert degrees to radians
    return math.sin(radians)

# Example usage
angle = float(input("Enter the angle in degrees: "))
print(f"sin({angle}) = {sin_theta(angle)}")
```

```python
import pandas as pd
from PIL import Image, ImageFilter, ImageTransform
import matplotlib.pyplot as plt
from skimage import feature, color
import numpy as np
```

```python
img = Image.open('cameraman.png')
img
```

#### Applying gray scale transformation

```python
gray_img = img.convert('L')
# Display the grayscale image
plt.imshow(gray_img, cmap='gray')
plt.axis('off')  # Hide axis
plt.show()
```

#### Apply Median Blur Filter

```python
blurred_img = img.filter(ImageFilter.MedianFilter(size=3))
plt.imshow(blurred_img, cmap='gray')
plt.axis('off')
plt.show()
```

#### Apply gaussian blur filter

```python
blurred_img = img.filter(ImageFilter.GaussianBlur(radius=2))
plt.imshow(blurred_img, cmap="gray")
```

#### Apply edge detection (canny filter)

```python
# Convert image to NumPy array
img_array = np.array(img)
# Apply Canny Edge Detection
edges = feature.canny(img_array, sigma=1.0)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```

#### Sobel Edge Detection

```python
from scipy.ndimage import sobel
```

```python
img_array = np.array(img)
# Apply Sobel filters (horizontal and vertical)
sobel_x = sobel(img_array, axis=0)  # Detects vertical edges
sobel_y = sobel(img_array, axis=1)  # Detects horizontal edges
sobel_combined = np.hypot(sobel_x, sobel_y)  # Computes gradient magnitude
sobel_combined = (sobel_combined / sobel_combined.max()) * 255  # Normalize to 0-255
plt.figure(figsize=(6, 6))
plt.imshow(sobel_combined, cmap="gray")
plt.title("Sobel Edge Detection (Combined)")
plt.axis("off")
plt.show()
```

#### Rotate the image by 90 degrees clockwise

```python
rotated_img = img.rotate(-90, expand=True)
plt.imshow(rotated_img, cmap='gray')
plt.axis('off')
plt.show()
```

#### Scale the image by factor of 1.5

```python
width, height = img.size
new_width = int(width * 1.5)
new_height = int(height * 1.5)
scaled_img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)
plt.imshow(scaled_img, cmap='gray')
plt.axis('off')
plt.show()
```

#### Translate the image with x-axis- 0.2 and y-axis-0.1

```python
width, height = img.size
tx = 0.2 * width  # 20% shift in x-axis
ty = 0.1 * height  # 10% shift in y-axis
translated_img = img.transform((width, height), ImageTransform.AffineTransform((1, 0, tx, 0, 1, ty)))
plt.imshow(translated_img, cmap='gray')
plt.show()
```

#### Enhance edge and find edges

```python
enhance = img.filter(ImageFilter.EDGE_ENHANCE_MORE)
edge = enhance.filter(ImageFilter.FIND_EDGES)
edge
plt.imshow(edge, cmap='gray')
```

#### Negative version of image

```python
imgarr = plt.imread('cameraman.png')
img_neg = 255 - imgarr
plt.imshow(img_neg, cmap='gray')
```

#### Flip the image horizontally

```python
# Flip the image horizontally
flipped_img = img.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(flipped_img, cmap="gray")
```

```python
width, height = img.size
new_size = (int(width * 5), int(height * 5))
# Resize with BICUBIC interpolation (good for up-sampling)
zoomed_img = img.resize(new_size, Image.BICUBIC)
plt.imshow(zoomed_img, cmap="gray")
```

```python
import scipy.ndimage
img_array = np.array(img.convert("L"))
# Apply zoom with SciPy interpolation (factor of 5)
zoomed_array = scipy.ndimage.zoom(img_array, zoom=5, order=3)
plt.imshow(zoomed_array, cmap="gray")
```
