```python
import sounddevice as sd
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
from scipy import interpolate
from scipy.signal import resample
```

```python
# Parameters for recording
duration = 30  # seconds
fs = 44100  # Sample rate (standard quality for audio)
```

```python
print("Recording...")
# Record audio
voice_recording = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype=np.int16)
sd.wait()  # Wait until recording is finished
print("Recording finished!")
```

```python
#wav.write('my_voice.wav', fs, voice_recording)
```

```python
print("Playing recorded audio...")
sd.play(voice_recording, fs)
sd.wait()  # Wait until the audio finishes playing
print("Playback finished!")
```

```python
# Print the shape of the signal
print("Shape of the recorded signal:", voice_recording.shape)
```

```python
# Convert to a 1D vector
voice_1d = voice_recording.flatten()
print("Shape of the 1D signal:", voice_1d.shape)
```

```python
# Plot the signal
plt.figure(figsize=(10, 4))
plt.plot(voice_1d)
plt.title("Recorded Speech Signal")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Take the first 15 seconds of the signal
duration = 15  # first 15 seconds
samples_15s = duration * fs  # Number of samples for 15 seconds
voice_first_15s = voice_recording[:samples_15s]

# Flatten the signal for plotting
voice_1d_15s = voice_first_15s.flatten()

# Plot the first 15 seconds of the signal
plt.figure(figsize=(10, 4))
plt.plot(voice_1d_15s)
plt.title("Recorded Speech Signal (First 15 Seconds)")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Downsample by a factor of 2 (Pick one sample for every two samples)
voice_downsampled = voice_first_15s[::2]

# Flatten the downsampled signal for plotting
voice_1d_downsampled = voice_downsampled.flatten()

# Plot the downsampled signal
plt.figure(figsize=(10, 4))
plt.plot(voice_1d_downsampled)
plt.title("Downsampled Speech Signal (First 15 Seconds, Downsampled by 2)")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()
new_fs = fs // 2  # New sampling rate after downsampling
# Play the downsampled audio
print("Playing the downsampled audio...")
sd.play(voice_downsampled, new_fs)
sd.wait()  # Wait until playback finishes
print("Playback finished!")
```

```python
samples_15s = duration * fs  # Number of samples for 15 seconds
voice_first_15s = voice_recording[:samples_15s].flatten()  # Flattening the 2D array to 1D

# Upsample by a factor of 2 (Insert new samples between existing samples)
n_original = len(voice_first_15s)
n_upsampled = 2 * n_original - 1  # New length after upsampling
x_original = np.arange(n_original)  # Original sample indices
x_upsampled = np.linspace(0, n_original - 1, n_upsampled)  # New indices for upsampling

# Linear interpolation to fill in new sample values
interpolator = interpolate.interp1d(x_original, voice_first_15s, kind='linear')
voice_upsampled = interpolator(x_upsampled)

# Plot the upsampled signal
plt.figure(figsize=(10, 4))
plt.plot(voice_upsampled)
plt.title("Upsampled Speech Signal (First 15 Seconds, Upsampled by 2)")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Flatten the signal for easier processing
voice_signal = voice_recording.flatten()

### 1. Downsampling by 2 ###
voice_downsampled = voice_signal[::2]  # Take one sample for every two samples
new_fs_downsampled = fs // 2  # New sample rate after downsampling

### 2. Upsampling by 2 ###
n_original = len(voice_signal)
n_upsampled = 2 * n_original - 1
x_original = np.arange(n_original)
x_upsampled = np.linspace(0, n_original - 1, n_upsampled)

# Linear interpolation to insert new samples
interpolator = interpolate.interp1d(x_original, voice_signal, kind='linear')
voice_upsampled = interpolator(x_upsampled)

new_fs_upsampled = fs * 2  # New sample rate after upsampling

# Save and play the upsampled audio
wav.write('my_voice_upsampled.wav', new_fs_upsampled, voice_upsampled.astype(np.int16))
print("Playing the upsampled audio...")
sd.play(voice_upsampled, new_fs_upsampled)
sd.wait()

### 3. Time-stretching (Speed Up) ###
# Resample to change the speed
factor = 1.5  # Speed up by 1.5x (faster)
n_resampled = int(len(voice_signal) / factor)  # Calculate the new number of samples
voice_timeshifted = resample(voice_signal, n_resampled)

# Save and play the time-stretched (speed-up) audio
wav.write('my_voice_timeshifted.wav', fs, voice_timeshifted.astype(np.int16))
print("Playing the time-stretched audio (faster)...")
sd.play(voice_timeshifted, fs)
sd.wait()

# Plot original and transformed signals (optional for visualization)
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(voice_signal)
plt.title('Original Speech Signal')

plt.subplot(3, 1, 2)
plt.plot(voice_downsampled)
plt.title('Downsampled Speech Signal')

plt.subplot(3, 1, 3)
plt.plot(voice_upsampled)
plt.title('Upsampled Speech Signal')

plt.tight_layout()
plt.show()
```
