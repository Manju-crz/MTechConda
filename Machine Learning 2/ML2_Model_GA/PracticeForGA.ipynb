```python
plt.figure(figsize=(8,5))
plt.boxplot(data['F3'])
plt.title('Boxplot of F3')
plt.ylabel('F3')
plt.show()
```

```python
numerical_cols = data.select_dtypes(include=['number']).columns.tolist()
print(numerical_cols)
```

```python
for column in numerical_cols:
    plt.figure(figsize=(8,5))
    plt.boxplot(data[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.show()
```

```python
fig, (a, b) = plt.subplots(1, 2, figsize=(20, 6))

sns.boxplot(y='F2', x='Skin_treatment', data=data, ax=a)
a.set_title('Boxplot of F2 by Skin Treatment')

sns.boxplot(y='F3', x='Skin_treatment', data=data, ax=b)
b.set_title('Boxplot of F3 by Skin Treatment')

plt.tight_layout()
plt.show()
```

```python
corr = data.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Choose your target column name
target_col = 'F29'  # Replace with your actual target column

# Compute correlation with numeric columns
corr_with_target = data.corr(numeric_only=True)[target_col].drop(target_col).sort_values(ascending=False)

# Print the correlation values
print(corr_with_target)

# Optional: Plot as a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=corr_with_target.values, y=corr_with_target.index, palette="coolwarm")
plt.title(f'Correlation of Features with "{target_col}"')
plt.xlabel('Correlation')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```

```python
selected_cols = ['F21', 'F15', 'F11', 'F24', 'F19', 'F12', 'F23', 'F8', 'F4', 'F28', 'F9', 'F16']  # Replace with your actual column names

# Subset the DataFrame
subset = data[selected_cols]

# Compute correlation matrix
corr_matrix = subset.corr(numeric_only=True)

# Print the correlation matrix
print(corr_matrix)

# Optional: Plot as heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix for Selected Features')
plt.tight_layout()
plt.show()
```
